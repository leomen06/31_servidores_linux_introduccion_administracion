Estados de un proceso:

![](img_128.png)

_ **Running or runnable (R)**: procesos que al ejecutarse se encuentran consumiendo recursos (de memoria o CPU). Cuando no esta en este estado, generalmente debido a que los recursos que el proceso requiere no est√°n disponibles o est√°n siendo usados por otro proceso, se dice que el proceso esta ‚Äòdurmiendo‚Äô (sleep), pero hay dos tipos de estados de ‚Äòdormir‚Äô o de ‚Äòsleep‚Äô: interrumpible y no interrumpible. El proceso sale del estado de ‚Äòsue√±o‚Äô al tener acceso a los recursos que necesita para continuar ejecut√°ndose.

_ **Uninterruptible Sleep (D)**: Este tipo de proceso no se pueden matar con una se√±al simple, ya que debe usarse una ‚Äòsign kill‚Äô para matar el proceso y que corresponde a un tipo de se√±al en particular.

_ **Interruptible Sleep (S)**: Son procesos que pueden interrumpirse con se√±ales normales siempre y cuando se encuentren en espera. Es importante recordar que las se√±ales son mecanismos que permiten al sistema terminar con procesos de forma ‚Äòamigable‚Äô y sin repercusiones para su funcionamiento.

_ **Stopped (T)**: Se refiere a procesos que han sido detenidos temporal o indefinidamente por un usuario. Se diferencia del estado en ‚Äòsleep‚Äô en el hecho de que un proceso que esta durmiendo sale de este estado al detectar que los recursos que necesita ya est√°n disponibles. En definitiba el proceso no saldr√° del estado Stopped hasta que se lo indiquemos.

_ **Zombie (Z)**: Es un estado correspondiente a los proceso que est√°n desvinculados de su proceso ‚Äòpadre‚Äô y que no est√°n ejecut√°ndose.

### <a name='mark_26'></a>Creaci√≥n y manejo de demonios
### [Indice](#index)

_ Los demonios es un mecan√≠smo que tiene linux para poder darle un comportamiento, especificamente llamado "de servicio", a los procesos, podemos inciarlos al principio del sistema o un determinado momento con ciertas condiciones, estos servicios se crean a trav√©z de un mecanismo que se llama Unit-file.

**En definitiva basicamente tendremos un archivo en alg√∫n lenguage el cual queremos que se ejecute constantemente y realice un determinada tarea.**

El primer demonio llamado `systemd` es el encargado (el primer demonio que se ejecuta en linux), el cual nos va a permitir controlar diferentes aspectos y servicios del sistema, inicia los drivers de red, el entorno gr√°fico, y todos los servicios que se ejecutan continuamente para hacer que el sistema operativo funciones.

Si nosotros queremos crear un demonio tenemos que agregarlo al entorno de "systemd", systemd tiene que estar buscando o viendo que los demonios exitan para poder agregarlos y ejecutarlos. Sin embargo systemd es tan complejo que solo se encarga del manejo de los demonios, no tiene una interfaz para poder controlarlos en si mismo. Para eso exites "systemctl" que permite manejar "el manejador de sistema" que tengamos.

**Systemd** crea y controla los demonios --> **systemctl** controla Systemd

_ systemd : crea los demonios

_ systemctl: gestiona los demonios

## Para crear un demonio primero debes:

Crear el script o unit file que usar√° de base tu demonio, esto puedes hacerlo con Python u otro lenguaje de scripting.

_ El siguiente script escribe en un archivo llamado timestamp.txt con un datetime por l√≠nea, cada segungo "time.sleep(1)", recordar que debemos tener creado el directorio "datelogs/" a nivel de root, (para que est√© disponible para todos los usuarios).

```python
import time
from datetime import datetime

while True:
    with opne("/datelogs/timestamp.txt", "a") as f:
        f.write("\nTimestamp: " + str(datetime.now()))
        f.close
    time.sleep(1)
```
_ Si quiere que este script se ejecute cada vez que se inicie el sistema, tengo que ponerlo como un "servicio".

_ Como root, vamos a `/etc/systemd/system`, en esta carpeta tenemos todos los archivos que se ejecutan como demonios, si exploramos un archivo ".service" al azar veremos la estructura en la que se escriben los Unit Files, archivos que usa systemd para saber como hacer tareas rutinarias, existen distintos tipos de archivos Unit Files aparte de los ".service".  

![](img_137.png)

_ Creamos el archivo ".service" con el comando `vim loggerpython.service` con el siguiente contenido.

```
[Unit]
Description=A python date logger
After=multi-user.target

[Service]
Type=simple
Restart=Always
ExecStart=/usr/bin/python3 /scripts/logger.py

[Install]
WantedBy=multi-user.target
```

_ El archivo cuenta con 3 secciones: 
    
    _ [Unit]:
    
        _ Contiene la descripcion, y "after"--> a partir de que momente se va a ejecutar, para este caso usamos "multi-user.target", que es una referencia que tiene "systemd" que una vez iniciados todos los servicios esenciales como, red, drivers, y diferentes funcionalidades de nuestro sistema operativo que nos permita acceder a una consola de texto. Tambi√©n son llamados runslevel, servicios que se inicializan al momento que se prende lo que es el systemd.
    
    _ [Service]:
    
        _ Type: tipo de servicio 
        _ Restart: Cuando se va a reiniciar.
        _ ExecStart: Comando que se ejecutar√° para realizar la tarea del demonio.
        
        
    _ [Install]: Una ves que ya se obtiene la especificaci√≥n del Unit y Service, en que momento lo voy a iniciar o a partir de que secci√≥n
    
Documentaci√≥n: [https://www.digitalocean.com/community/tutorials/understanding-systemd-units-and-unit-files](https://www.digitalocean.com/community/tutorials/understanding-systemd-units-and-unit-files)

_ Guardamos los cambios, ya tenemos nuestro demonio, pero para que nuestro demonio pueda correr necesitamos reiniciar systemd con `systemctl deamon-reload`.

_ Ahora checkeamos el estado del servicio que creamos con `systemctl status nombre_servicio`, como vemos el servicio esta "muerto" y tenemos que activarlo manualmente.
